```
# импортируем модули в тетрадку

import requests as rq

from bs4 import BeautifulSoup
```

```
asahi2 = "https://sitesearch.asahi.com/sitesearch/index.php?Keywords=%E9%95%B7%E5%B4%8E&tvsubmit=%E6%A4%9C%E7%B4%A2&iref=pc_ss_date"
```

```
asahi_page1 = rq.get(asahi1)
asahi_page1
```

```
asahi_page2 = rq.get(asahi2)
asahi_page2
```

```
asahi_page1.text
```

```
asahi_soup1 = BeautifulSoup(asahi_page1.text, 'html.parser') #сохраним результат в переменную soup
```

```
print(asahi_soup1.prettify()) # показывает нашу страницу в красивом виде
```

```
asahi_soup1.find_all('a') [7].get_text()
```

```
links = []
soup1list1 = asahi_soup1.find_all('a')
#soup1list1
```

```
for elem in soup1list1: # для элемента списка
  links.append(elem.get_text()) # добавляем нужный атрибут(ссылку) в список для ссылок
links
```

```
links_lower = []
for elem in links:
  elem = elem.lower()
  if elem != '':
    links_lower.append(elem)
links_lower
```
