```
def ngrams_by_nltk(word, num):
  word = word.lower()
  word = re.sub("[!\"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~]?","",word)
  tokens = word.split() # сделать с местоимениями и частотностью Я #  здесь задаем паттерн
  tokens = re.findall("я\s\w+", word)
  output = list(ngrams(tokens, num)) 
  return output
```

```
def ngrams_by_nltk(word, num):
  word = word.lower()
  word = re.sub("[!\"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~]?","",word)
  tokens = word.split() # сделать с местоимениями и частотностью Я #  здесь задаем паттерн
  tokens = re.findall("(.{2,5}(ил|ул|ал))", word)
  output = list(ngrams(tokens, num)) 
  res = Counter(output).most_common()

  return res
```
