```
# импортируем модули в тетрадку

import requests as rq

from bs4 import BeautifulSoup
```

```
asahi1 = "http://sitesearch.asahi.com/.cgi/ajwsitesearch/sitesearch.pl?Keywords=hiroshima"
```

```
asahi2 = "http://sitesearch.asahi.com/ajwsitesearch/?Keywords=nagasaki&tvsubmit=Search"
```

```
asahi_page1 = rq.get(asahi1)
asahi_page1
```

```
asahi_page2 = rq.get(asahi2)
asahi_page2
```

```
asahi_page1.text
```

```
asahi_soup1 = BeautifulSoup(asahi_page1.text, 'html.parser') #сохраним результат в переменную soup
```

```
print(asahi_soup1.prettify()) # показывает нашу страницу в красивом виде
```
