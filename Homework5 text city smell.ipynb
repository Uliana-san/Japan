```
my_file = open('city_smells.txt', 'r')
v_txt = my_file.read()
print(v_txt)
```

```
# записать файл пока не смогла, тк пропустила пару и не разобралась с write
# высылаю код и результат
# посчитать предложения не поняла, как - может, потому, что сразу убрала знаки препинания и исчезло разделение на предложения?

v_txt_a = v_txt.lower().replace(","," ").replace("."," ").replace("?"," ").replace("!"," ").replace(":"," ").replace("("," ").replace(")"," ")    #strip("\n").strip(".,?!:-1234567890")
#print(v_txt_a)
v_a = v_txt_a.split()
v_s = set(v_a)
v_len = 0
for x in v_s:
  v_len = v_len + len(x)

v_len_a = 0
for x in v_a:
  v_len_a = v_len_a + len(x)

print("Общая длина всех уникальных слов",v_len,"Количество уникальных слов",len(v_s),"Общее количество слов",len(v_a),"Общая длина всех слов",v_len_a)
print("Средняя длина уникального слова", v_len/len(v_s))
print("Средняя длина слова", v_len_a/len(v_a))
print ("Отношение количества уникальных слов к общему количеству слов", len(v_s)/len(v_a))
```
