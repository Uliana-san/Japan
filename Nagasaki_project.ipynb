```
links = []
soup1list2 = ria_soup2.find_all('a')
#soup1list2
```

```
from pymorphy2 import MorphAnalyzer
from pymystem3 import Mystem
```

```
mystem = Mystem() 
morph = MorphAnalyzer() 
```

```
print(ria2_str_clean)
print (mystem.lemmatize(ria2_str_clean))
```

```
words_analized2 = mystem.analyze(ria2_str_clean)
words_analized2
```

```
res_nagasaki =[]
for parse in words_analized2:
  if parse.get("analysis"):
    res_nagasaki.append(parse["analysis"][0]["lex"])
res_nagasaki
```

```
#stopwords for russian
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
russian_stopwords = stopwords.words("russian")

from string import punctuation
```

```
for word in res_nagasaki:
  if word not in russian_stopwords:
    print (word)
```

```
res_nagasaki_final = []
for word in res_nagasaki:
  if word not in russian_stopwords:
    res_nagasaki_final.append(word)
res_nagasaki_final
  
```

```
from collections import defaultdict as dd
from operator import itemgetter
```

```
# wordcloud stuff

import matplotlib.pyplot as plt
from wordcloud import WordCloud

def show_wordcloud_by_freq_dict(freq_dict):
    wordcloud = WordCloud()
    wordcloud.generate_from_frequencies(frequencies=freq_dict)
    plt.figure()
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```

```
text_words_frequencies = dd(int) # это копируем
        
for word in res_nagasaki_final:
  text_words_frequencies[word] += 1 / len(res_nagasaki_final)

sorted_frequency_table = sorted(text_words_frequencies.items(), 
                                        key=itemgetter(1), reverse=True)    
        
for word, freq in sorted_frequency_table[:20]:
  print('\t'.join((word, str(freq))))
        
show_wordcloud_by_freq_dict(text_words_frequencies)
print(res_nagasaki_final)
```
