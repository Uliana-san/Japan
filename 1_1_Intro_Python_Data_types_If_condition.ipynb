Формально, каждая переменная в языке Python состоит из следующих частей:
* **имя** переменной (то имя, которое мы для нее придумываем на этапе создания)
* **тип** (тип данных, которые хранятся в этой переменной)
* **значение** (непосредственно само значение, которое было присвоено переменной)

Python — язык с **динамической типизацией**. Это значит, что вам не нужно явно объявлять тип переменных, тип определяется автоматически при присвоении значения.

Рассмотрим **основные типы данных:**

* Числа:
    * **int** — целые числа (напр., 0, 11, -1500)
    * **float** — нецелые числа или числа с плавающей точкой (напр., 0.0, 3.14, -99.999)
    * **complex** — комплексные числа (напр., 3 + 2j)
* Строки:
    * **str** — последовательность символов, заключенная в кавычки (напр., "Hello world", 'Привет')
* Логический тип:
    * **bool** — переменные этого типа могут принимать всего два значения: **True** (истина) или **False** (ложь)
* Списки:
    * **list** —  список (напр., [1, 2, 3])
    * **tuple** — кортеж (напр., (1, 2, 3))
    * **range** — диапазон
* None:
    * **None** — отдельный тип данных, имеющий единственное значение None, предназначенный для пустых (отсутствующих) значений
* Словари:
    * **dict** — тип данных, имеющий структуру словаря
* Множества:
    * **set** — тип данных, схожий по своей организации с математическим множеством
    * **frozenset** — неизменяемое множество

### 2.2. Числовые типы `int` и `float`

**int** (целые числа):  
Примеры: `10`, `-5`, `0`.  
Преимущества:
- В Python `int` не ограничен 32 или 64 битами, как во многих других языках. Число может быть очень большим, пока хватает памяти.

**float** (вещественные числа):  
Примеры: `3.14`, `-7.5`, `2.0`.  
Особенности:
- `float` в Python обычно 64-битный (двойная точность).  
- При выполнении сложных вычислений могут возникать некоторые неточности, связанные с представлением дробных чисел

Ниже — простые примеры арифметических операций.

```python
# ?range
```

```python
a = 10
b = 3

print("a =", a)
print("b =", b)
print("Сложение (a + b):", a + b)
print("Вычитание (a - b):", a - b)
print("Умножение (a * b):", a * b)
print("Деление (a / b):", a / b) # результат float
print("Целочисленное деление (a // b):", a // b)
print("Остаток от деления (a % b):", a % b)
print("Возведение в степень (a ** b):", a ** b)
```

Обратите внимание, что оператор `/` возвращает `float`, даже если результат кажется целым числом.

По ходу работы программы переменная может принимать различные значения. Код обрабатывается построчно **сверху вниз**.

Если в строке выше присвоено определенное значение, то его можно использовать в строках ниже. Если значение переменной было изменено, то во всех строках ниже будет использоваться уже новое значение.

Знак **присвоения ( = )** всегда обрабатывается **справа налево**, то есть как только встречается строка, содержащая данный символ, то всегда вычисляется значение справа от знака равно и присваивается переменной, указанной слева от знака равно.

### Преобразование типов

Самыми частыми являются следующие преобразования:<br>
**int <-> float** (целое число <-> нецелое число)<br>
**int <-> str** (целое число <-> строка)<br>
**float <-> str** (нецелое число <-> строка)

```python
number_int = 55
print(number_int)
print(type(number_int))
```

```python
number_float = float(number_int)
print(number_float)
print(type(number_float))
```

```python
number_int_2 = int(number_float)
print(number_int_2)
print(type(number_int_2))
```

```python
a = 55.9999999

b = int(a)

print(b)
```

```python
# преобразование целочисленной переменной к нецелочисленному и строковому типам

# исходная переменная типа int
number_int = 55

# создаем новую переменную типа float со значением старой переменной
number_float = float(number_int)

# создаем новую переменную типа str со значением старой переменной
number_str = str(number_int)

# выводим на печать значения переменных и их типы
print(number_int, type(number_int))
print(number_float, type(number_float))
print(number_str, type(number_str))
```

При преобразовании **нецелого числа к целому** дробная часть (при ее наличии) **отбрасывается**. Округления при этом не происходит!

```python
number_1 = 55.8
print(type(number_1))
```

```python
value_1 = int(number_1)

print(value_1)
print(type(value_1))
```

```python
# преобразование нецелого числа к целому

# исходные переменные с числами float
number_1 = 55.0

# получаем переменные типа int
value_1 = int(number_1)

# выводим на печать значения переменных и их типы
print(value_1, type(value_1))
```

Если величина имеет исходно **строковый тип**, ее **можно преобразовать к числу**. <br>
Но это возможно только в том случае, когда данное преобразование допустимо (то есть, когда строка содержит такую последовательность символов, которая в самом деле могла бы быть числом).<br>
При попытке преобразования к числу некорректных строк программа выдаст **ошибку**.

### 2.4. Булевы значения `bool`

Булевы значения могут принимать `True` или `False`. Они возникают при сравнении, логических операциях и т.д.

**Операторы сравнения**:
- `==` (равенство)
- `!=` (неравенство)
- `>`, `<`, `>=`, `<=`

**Логические операторы**:
- `and` - логическое умножение
- `or` - логическое сложение
- `not` - логическое отрицание

```python
5 == 5
```

```python
x = 5
y = 10

print(not(x == 4))
```

### Задание 5.4 Манхэттонское растояние

В машинном обучении и анализе данных вычисление расстояния между точками часто используется для оценки сходства объектов. Например, в алгоритмах кластеризации (`k-means`) необходимо определить насколько объекты близки друг к другу по своим признакам, что позволяет группировать данные или находить наиболее похожие элементы.

**Манхэттенское расстояние** (или расстояние городских кварталов) — это способ измерения расстояния между двумя точками на плоскости, при котором перемещение осуществляется только по вертикали и горизонтали (как по сетке улиц на карте города). Название происходит от района Манхэттен в Нью-Йорке, где улицы образуют прямоугольную решётку.

Для двух точек $$A(x_1, y_1)$$ и $$B(x_2, y_2)$$ манхэттенское расстояние вычисляется по формуле:

$$ d(A, B) = |x_1 - x_2| + |y_1 - y_2| $$

То есть расстояние равно сумме модулей разностей координат по каждой из осей.

### Задание 5.5 Евклидово расстояние

Евклидово расстояние — это привычное нам расстояние «по прямой» между двумя точками на плоскости. Оно вычисляется по теореме Пифагора и показывает длину кратчайшего пути, соединяющего точки.

На плоскости евклидово расстояние между двумя точками $(x_1; y_1)$ и $(x_2; y_2)$ определяется формулой:

$$ d = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2} $$
