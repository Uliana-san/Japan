```python
import pandas as pd
import re
from collections import Counter
import matplotlib.pyplot as plt
!pip install docx2txt
!pip install stanza
import stanza
import docx2txt
#stanza.download("ja")
nlp_stanza = stanza.Pipeline(lang="ja", processors="tokenize, pos, lemma, depparse, ner")
```

```python
kisida_text = docx2txt.process("/content/岸田首相所信発明.docx")
hatoyama_text = docx2txt.process("/content/鳩山内閣総理大臣所信表明演説.docx")
```

```python
doc_kisida = nlp_stanza(kisida_text)
doc_hatoyama = nlp_stanza(hatoyama_text)
```

```python
list_of_rows = []
counter = 0
for sentence in doc_kisida.sentences:
    counter += 1
    for word in sentence.words:
        list_of_rows.append([counter, word.id, word.text, word.upos, word.deprel, word.head, sentence.words[word.head-1].text])

df_kisida = pd.DataFrame(list_of_rows, columns=['sent_id', 'id', 'token', 'pos', 'synt_tag', 'head_id', 'head_tok'])
df_kisida[:10]
```

```python
list_of_rows = []
counter = 0
for sentence in doc_hatoyama.sentences:
    counter += 1
    for word in sentence.words:
        list_of_rows.append([counter, word.id, word.text, word.upos, word.deprel, word.head, sentence.words[word.head-1].text])

df_hatoyama = pd.DataFrame(list_of_rows, columns=['sent_id', 'id', 'token', 'pos', 'synt_tag', 'head_id', 'head_tok'])
df_hatoyama[:10]
```

```python
adv_k = df_kisida.loc[(df_kisida['pos'] == 'ADV')]
adv_k[:10]
```

```python
adv_h = df_hatoyama.loc[(df_hatoyama['pos'] == 'ADV')]
adv_h[:10]
```

```python
adv_k_list = adv_k['token'].tolist()
adv_h_list = adv_h['token'].tolist()
```

```python
import numpy as np
```

```python
pd.value_counts(np.array(adv_k_list))
```

```python
pd.value_counts(np.array(adv_h_list ))
```
