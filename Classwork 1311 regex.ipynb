```
import re
```

```
result = re.match(r'Hello','Hello my dear friend Amy')
print(result)
```

```
print(result.group(0))
```

```
result = re.search(r'my','Hello my dear friend Amy')
print(result.group(0))
```

```
result = re.fullmatch(r'Hello my dear friend Amy', 'Hello my dear friend Amy')
print (result)
print ("\n", result.group(0)) # можно без н
```

```
result = re.findall(r'my','Hello my dear friend Amy')
print(result)
```

```
result = re.sub(r'Amy','Jack','hello my dear friend Amy. How are you, Amy?')
print(result)
```

```
result = re.split(r'e','Hello my dear friend Amy')
print(result)
```

```
# сделать разбивку по предложениям
```

```
result = re.split(r'\.','hello my dear friend Amy.I\'m so happy to see you.')
print(result)
```

```
result = re.split(r'e','hello me dear friend Amy, eee eee',maxsplit=2)
print(result)
```

```
result = re.match(r'\d+','1984 hello')
print(result.group())
```
