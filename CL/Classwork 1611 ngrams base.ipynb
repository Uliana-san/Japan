```
def ngrams_by_nltk(word, num):
  word = word.lower()
  word = re.sub("[!\"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~]?","",word)
  tokens = word.split() #       #    
  tokens = re.findall("\s\w+", word)
  output = list(ngrams(tokens, num)) 
  return output
```

```
def ngrams_by_nltk(word, num):
  word = word.lower()
  word = re.sub("[!\"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~]?","",word)
  tokens = word.split() #       #    
  tokens = re.findall("(.{2,5}(||))", word)
  output = list(ngrams(tokens, num)) 
  res = Counter(output).most_common()

  return res
```
